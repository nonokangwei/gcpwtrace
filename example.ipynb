```python
credentials = service_account.Credentials.from_service_account_file(
    '/usr/local/google/home/hzchen/sa-dave-selfstudy01.json',
)

sql = """
    SELECT * FROM `dave-selfstudy01.wtrace.tx_result`
    WHERE
        dest_url = 'http://8.219.85.89/gcp_gaming.pdf'
    OR
        dest_url = 'http://8.209.247.102/gcp_gaming.pdf'
    OR
        dest_url = 'http://47.254.134.72/gcp_gaming.pdf'
    OR
        dest_url = 'http://34.124.217.36/gcp_gaming.pdf'
    OR
        dest_url = 'http://34.85.102.60/gcp_gaming.pdf'
    OR
        dest_url = 'http://34.159.107.227/gcp_gaming.pdf'
    OR
        dest_url = 'http://35.204.234.20/gcp_gaming.pdf'
    AND response_code = 200
    """
df = pandas_gbq.read_gbq(sql, project_id="dave-selfstudy01", credentials=credentials, dialect="standard")
df.head()
```

```python
def country_code2name(code):
    dictionary = dict({
        "ID": "印尼",
        "MY": "马来西亚",
        "TH": "泰国",
        "VN": "越南",
        "BN": "文莱",
        "PH": "菲律宾",
        "MM": "缅甸",
        "LA": "老挝",
        "KH": "柬埔寨",
        "SG": "新加坡",
        "JP": "日本",
        "KR": "韩国",
        "HK": "香港",
        "TW": "台湾",
        "MO": "澳门",
        "RU": "俄罗斯",
        "GB": "英国",
        "CH": "瑞士",
        "ES": "西班牙",
        "NL": "荷兰",
        "IT": "意大利",
        "SE": "瑞典",
        "FR": "法国",
        "GE": "德国",
        "PL": "波兰"})
    if code in dictionary.keys():
        return dictionary[code]
    else:
        return code

df['country'] = df.country_code.apply(country_code2name)
```

```python
df['vendor'] = ''
df['region'] = ''
df.loc[df.dest_url.str.contains('8.219.85.89'), 'region'] = 'singapore'
df.loc[df.dest_url.str.contains('8.209.247.102'), 'region'] = 'tokyo'
df.loc[df.dest_url.str.contains('47.254.134.72'), 'region'] = 'frankfurt'
df.loc[df.dest_url.str.contains('34.124.217.36'), 'region'] = 'singapore'
df.loc[df.dest_url.str.contains('34.85.102.60'), 'region'] = 'tokyo'
df.loc[df.dest_url.str.contains('34.159.107.227'), 'region'] = 'frankfurt'
df.loc[df.dest_url.str.contains('35.204.234.20'), 'region'] = 'netherland'

df.loc[df.dest_url.str.contains('8.219.85.89'), 'vendor'] = 'ali'
df.loc[df.dest_url.str.contains('8.209.247.102'), 'vendor'] = 'ali'
df.loc[df.dest_url.str.contains('47.254.134.72'), 'vendor'] = 'ali'
df.loc[df.dest_url.str.contains('34.124.217.36'), 'vendor'] = 'gcp'
df.loc[df.dest_url.str.contains('34.85.102.60'), 'vendor'] = 'gcp'
df.loc[df.dest_url.str.contains('34.159.107.227'), 'vendor'] = 'gcp'
df.loc[df.dest_url.str.contains('35.204.234.20'), 'vendor'] = 'gcp'
```
